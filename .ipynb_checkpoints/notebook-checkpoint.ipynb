## Import Semua Packages/Library yang Digunakan

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Wrangling

### Gathering Data

```python
# Memuat tabel day menjadi sebuah DataFrame
day_df = pd.read_csv('./data/day.csv')
day_df.head()
```

```python
# Memuat tabel hour menjadi sebuah DataFrame
hour_df = pd.read_csv('./data/hour.csv')
hour_df.head()
```

### Assessing Data

```python
# Menilai data day_df
day_df.info()
```

```python
print("Jumlah duplikasi: ", day_df.duplicated().sum())
day_df.describe()
```

```python
# Menilai data hour_df
hour_df.info()
```

```python
print("Jumlah duplikasi: ", hour_df.duplicated().sum())
hour_df.describe()
```

**Insight:**
- Terdapat kesalahan tipe data pada column dteday. Column dteday berisi tanggal-bulan-tahun yang seharusnya memiliki tipe data datetime64.

### Cleaning Data

```python
datetime_column = ["dteday"]

# Untuk tabel day_df
for column in datetime_column:
    day_df[column] = pd.to_datetime(day_df[column])

# Untuk tabel hour_df
for column in datetime_column:
    hour_df[column] = pd.to_datetime(hour_df[column])
```

```python
day_df.info()
```

```python
hour_df.info()
```

# Mengubah angka menjadi keterangan pada day_df
# Pada tabel day, terdapat season, yr, mnth, holiday, weekday, workingday, dan weathersit yang dituliskan menggunakan angka 

```python
day_df['season'] = day_df['season'].map({
    1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'
})
day_df['yr'] = day_df['yr'].map({
    0: '2011', 1: '2012'
})
day_df['mnth'] = day_df['mnth'].map({
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6:'Jun', 7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
})
day_df['weekday'] = day_df['weekday'].map({
    0: 'Sun', 1: 'Mon', 2: 'Tue', 3: 'Wed', 4: 'Thu', 5: 'Fri', 6: 'Sat'
})
day_df['workingday'] = day_df['workingday'].map({
    0: 'Holiday', 1: 'Workingday'
})
day_df['weathersit'] = day_df['weathersit'].map({
    1: 'Clear/Partly Cloudy',
    2: 'Misty/Cloudy',
    3: 'Light Snow/Rain',
    4: 'Severe Weather'
})

day_df['season'] = day_df.season.astype('category')
day_df['yr'] = day_df.yr.astype('category')
day_df['mnth'] = day_df.mnth.astype('category')
day_df['holiday'] = day_df.holiday.astype('category')
day_df['weekday'] = day_df.weekday.astype('category')
day_df['workingday'] = day_df.workingday.astype('category')
day_df['weathersit'] = day_df.weathersit.astype('category')

day_df.head()
```

# Mengubah angka menjadi keterangan pada hour_df
# Pada tabel day, terdapat season, yr, mnth, holiday, weekday, workingday, dan weathersit yang dituliskan menggunakan angka 

```python
hour_df['season'] = hour_df['season'].map({
    1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'
})
hour_df['yr'] = hour_df['yr'].map({
    0: '2011', 1: '2012'
})
hour_df['mnth'] = hour_df['mnth'].map({
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6:'Jun', 7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
})
hour_df['weekday'] = hour_df['weekday'].map({
    0: 'Sun', 1: 'Mon', 2: 'Tue', 3: 'Wed', 4: 'Thu', 5: 'Fri', 6: 'Sat'
})
hour_df['workingday'] = hour_df['workingday'].map({
    0: 'Holiday', 1: 'Workingday'
})
hour_df['weathersit'] = hour_df['weathersit'].map({
    1: 'Clear/Partly Cloudy',
    2: 'Misty/Cloudy',
    3: 'Light Snow/Rain',
    4: 'Severe Weather'
})

hour_df['season'] = hour_df.season.astype('category')
hour_df['yr'] = hour_df.yr.astype('category')
hour_df['mnth'] = hour_df.mnth.astype('category')
hour_df['holiday'] = hour_df.holiday.astype('category')
hour_df['weekday'] = hour_df.weekday.astype('category')
hour_df['workingday'] = hour_df.workingday.astype('category')
hour_df['weathersit'] = hour_df.weathersit.astype('category')

day_df.head()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

Bagaimana pola penyewaan sepeda berdasarkan musim?

```python
season_colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red'] 

fig, ax = plt.subplots(1, 2, figsize=(14, 6))

sns.countplot(
    x='season',
    data=day_df,
    hue='season',
    palette=season_colors,
    legend=False,
    ax=ax[0]
)
ax[0].set_title('Distribusi Musim', fontsize=14, fontweight='bold')
ax[0].set_xlabel('Musim', fontsize=12)
ax[0].set_ylabel('Jumlah', fontsize=12)

sns.barplot(
    x='season',
    y='registered',
    data=day_df,
    hue='season',
    label='Terdaftar',
    palette=season_colors,
    legend=False,
    ax=ax[1],
    width=0.5,
    errorbar=None
)
sns.barplot(
    x='season',
    y='casual',
    data=day_df,
    hue='season',
    label='Kasual',
    palette=season_colors, 
    legend=False,
    hatch='/',
    ax=ax[1],
    width=0.5,
    errorbar=None
)
ax[1].set_title('Korelasi Pengguna Terdaftar dan Kasual', fontsize=14, fontweight='bold')
ax[1].set_xlabel('Musim', fontsize=12)
ax[1].set_ylabel('Rata-rata Jumlah Penyewa', fontsize=12)
ax[1].legend(['Terdaftar', 'Kasual'], loc='upper left')

plt.tight_layout()

plt.show()
```

**Insight:**
Pola penyewaan sepeda berdasarkan musim dipetakan ke dalam dua tabel, yaitu tabel distribusi musim yang menunjukkan jumlah frekuensi atau kejadian data, serta tabel korelasi pengguna terdaftar dan kasual yang menunjukkan rata-rata jumlah penyewa.

Dari tabel distribusi musim, frekuensi tertinggi tercatat pada musim gugur (Fall) dengan 188 data, sementara frekuensi terendah tercatat pada musim dingin (Winter) dengan 178 data.

Pada tabel korelasi pengguna, rata-rata jumlah penyewa terdaftar tertinggi ditemukan pada musim gugur (Fall) dengan 4441,691 penyewa, sedangkan yang terendah tercatat pada musim semi (Spring) dengan 2269,204 penyewa.

Rata-rata jumlah penyewa kasual tertinggi juga terjadi pada musim gugur (Fall) dengan 1202,611 penyewa, sementara yang terendah tercatat pada musim semi (Spring) dengan 334,928 penyewa.

### Pertanyaan 2:

Bagaimana pola penyewaan sepeda berdasarkan kondisi cuaca?

```python
weathersit_colors = ['tab:blue', 'tab:orange', 'tab:green']

fig, ax = plt.subplots(1, 2, figsize=(12, 7))

sns.countplot(
    x='weathersit', 
    data=day_df, 
    hue='weathersit',
    ax=ax[0], 
    palette=weathersit_colors, 
    width=0.5
)
ax[0].set_title('Distribusi Kondisi Cuaca', fontsize=14, fontweight='bold')
ax[0].set_xlabel('Kondisi Cuaca', fontsize=12, fontweight='medium')
ax[0].set_ylabel('Jumlah', fontsize=12, fontweight='medium')

sns.barplot(
    x='weathersit', 
    y='registered', 
    data=day_df, 
    label='Pengguna Terdaftar', 
    color='tab:blue', 
    ax=ax[1], 
    width=0.5, 
    errorbar=None
)
sns.barplot(
    x='weathersit', 
    y='casual', 
    data=day_df, 
    label='Pengguna Kasual',  
    color='tab:orange', 
    ax=ax[1], 
    width=0.5, 
    errorbar=None
)
ax[1].set_title('Korelasi Pengguna Terdaftar dan Kasual', fontsize=14, fontweight='bold')
ax[1].set_xlabel('Kondisi Cuaca', fontsize=12, fontweight='medium')
ax[1].set_ylabel('Rata-rata Jumlah Pengguna', fontsize=12, fontweight='medium')

ax[1].legend(title='Tipe Pengguna')

plt.tight_layout()
plt.show()
```

**Insight:**
Pola penyewaan sepeda berdasarkan cuaca dipetakan ke dalam dua tabel, yaitu tabel distribusi cuaca yang menunjukkan jumlah frekuensi atau kejadian data, serta tabel korelasi pengguna terdaftar dan kasual yang menunjukkan rata-rata jumlah penyewa.

Dari tabel distribusi cuaca, frekuensi tertinggi tercatat pada saat kondisi cuaca cerah (Clear/Partly Cloudy) dengan 463 data, sementara frekuensi terendah tercatat pada saat kondisi cuaca hujan (Light Snow/Rain) dengan 21 data.

Pada tabel korelasi pengguna, rata-rata jumlah penyewa terdaftar tertinggi ditemukan pada saat kondisi cuaca cerah (Clear/Partly Cloudy) dengan 3912,755 penyewa, sedangkan yang terendah tercatat pada saat kondisi cuaca hujan (Light Snow/Rain) dengan 1617,809 penyewa.

Rata-rata jumlah penyewa kasual tertinggi juga terjadi pada saat kondisi cuaca cerah (Clear/Partly Cloudy) dengan 964,030 penyewa, sementara yang terendah tercatat pada saat kondisi cuaca hujan (Light Snow/Rain) dengan 185,476 penyewa.

## Conclusion

- Konklusi dari pertanyaan 1: Bagaimana pola penyewaan sepeda berdasarkan musim?
Penyewaan sepeda menunjukkan pola yang bervariasi menurut musim. Musim gugur (Fall) tercatat sebagai musim dengan frekuensi data dan rata-rata jumlah penyewa terdaftar serta kasual tertinggi, baik untuk pengguna terdaftar maupun kasual. Sebaliknya, musim semi (Spring) memiliki jumlah penyewa terdaftar dan kasual terendah. Ini menunjukkan bahwa musim gugur mungkin menjadi periode puncak penyewaan sepeda, sementara musim semi cenderung memiliki jumlah penyewa yang lebih rendah.
Saran: sistem penyewaan sepeda dapat ditingkatkan dengan menawarkan promosi khusus pada musim semi untuk menarik lebih banyak pelanggan.
- Konklusi dari pertanyaan 2: Bagaiman pola penyewaan sepeda berdasarkan kondisi cuaca?
Penyewaan sepeda menunjukkan pola yang bervariasi berdasarkan kondisi cuaca. Cuaca cerah (Clear/Partly Cloudy) tercatat sebagai kondisi cuaca dengan frekuensi data dan rata-rata jumlah penyewa tertinggi, baik untuk pengguna terdaftar maupun kasual. Sebaliknya, cuaca hujan (Light Snow/Rain) memiliki frekuensi data dan rata-rata jumlah penyewa terendah. Ini menunjukkan bahwa cuaca cerah menjadi waktu puncak untuk penyewaan sepeda, sementara cuaca hujan cenderung mengurangi minat pengguna dalam menyewa sepeda.
Saran: sistem penyewaan sepeda dapat ditingkatkan dengan menawarkan promosi atau diskon khusus pada hari-hari dengan cuaca hujan untuk meningkatkan jumlah penyewa. Selain itu, penyedia jasa penyewaan dapat menyediakan fasilitas tambahan seperti jas hujan atau rute yang lebih aman untuk mendukung pengguna dalam kondisi cuaca kurang ideal.
